```python
import plot_core as pc
import tkinter as tt
import tkinter.filedialog
import os


def get_data():
    a = tkinter.filedialog.askopenfilename()
    f = open("temp", 'w')
    # write RECV file.
    f.write(a)

    a = a.strip(".RECV")
    a = a + ".RESP"
    f.write("\n")
    f.write(a)

    sitename = a.strip(".RESP")
    sitename = sitename.split("/")
    length = len(sitename)
    sitename = sitename[length-1] + ".RESP"
    path = a.strip(sitename)
    f.write("\n")
    f.write(path)
    f.close()


def plotdata():
    f = open("temp")
    recv_filename = f.readline()
    recv_filename = recv_filename.strip("\n")
    resp_filename = f.readline()
    resp_filename = resp_filename.strip("\n")
    filepath = f.readline()
    pc.mt1dinv_plot(recv_filename, resp_filename, filepath)
    f.close()
    os.system("rm temp")
    notice = "Figure saved to Result.png!"
    tkinter.messagebox.showinfo("Figure Saved", notice)


recv_filename = "/home/shenyao-jin/文档/2022春夏/地球电磁/term/mtinv_plot/IAJ34/IAJ34.RECV"
resp_filename = "/home/shenyao-jin/文档/2022春夏/地球电磁/term/mtinv_plot/IAJ34/IAJ34.RESP"

top = tt.Tk()
top.title("MT1DCC plot tool")
tkinter.messagebox.showinfo("Notice", "This version is base on shell, theoretically it could work on cygwin.")
tt.Label(top, text="MT1DOCC plot tool in Python", font=("Source Sans Pro", 16)).grid(row=1, column=1)
tt.Label(top, text="Shenyao Jin", font=("Source Sans Pro", 16)).grid(row=2, column=1)
tt.Label(top, text="Zhejiang Univ.", font=("Source Sans Pro", 16)).grid(row=3, column=1)

button = tt.Button(top, width=20, height=2, text='Choose .RECV', command=get_data)
button.grid(row=9)

button1 = tt.Button(top, width=20, height=2, text='Calculate', command=plotdata)
button1.grid(row=9, column=2)

top.mainloop()
```
